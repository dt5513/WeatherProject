```python
import pymongo
import pickle
import pandas as pd

connection = pymongo.MongoClient("mongodb://localhost")
db = connection.metoffice
collection = db.tweets_wt

data = db.tweets_wt.find()

stats = {}
baseCounters = {'sunny': 0, 'cloudy': 0, 'snow': 0, 'rain': 0, 'foggy': 0, 'none': 0, 'cold': 0, 'forecasts': 0, 'total':0}
for d in data:
    datetime = '-'.join([d['date'],str(d['t'])])
    if stats.get(datetime,'') == '':
        stats[datetime] = {'sunny': 0, 'cloudy': 0, 'snow': 0, 'rain': 0, 'foggy': 0, 'none': 0, 'cold': 0, 'forecasts': 0, 'total':0}
    stats[datetime]['total'] += 1
    stats[datetime][d['wt']] += 1

for s in stats:
     print s, stats[s]
datetime_column = ['date', 'time']
columns = ['none','sunny','cloudy','foggy','rain','snow','cold','forecasts','total']
csvlist = []

for s in stats:
    tempList = [0] * 2
    tempList[0] = s[0:10]
    tempList[1] = s[11:]
    for c in columns:
        tempList.append(stats[s][c])
    csvlist.append(tempList)
```

```python
wtDf = pd.DataFrame(stats).T
```

```python
datetimesoftweets = wtDf.index
dates = []
dates.extend(d[:10] for d in datetimesoftweets)
wtDf['date'] = dates
times = []
times.extend(d[11:] for d in datetimesoftweets)
wtDf['time'] = times
```
