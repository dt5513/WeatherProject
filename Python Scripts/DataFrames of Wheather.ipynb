```python
import pymongo, pickle
import datetime as dt
import pandas as pd

connection = pymongo.MongoClient("mongodb://localhost")
db = connection.metoffice
collection = db.weather_dt
data = db.weather.find()

rawData = []
for d in data:
    print d
    rawData.append(d)


```

```python
dictData = {}
for row in rawData:
    dayDiff = 0
    datetime = '-'.join([row['date'],str(row['t'])])
    if dictData.get(datetime,'') == '':
        dictData[datetime] = {}
        
    dayDiff = int(((dt.datetime.strptime(row['date'], '%Y-%m-%d') - dt.datetime.strptime(row['date obtained'], '%Y-%m-%d')).total_seconds() /86400))
    
    dictData[datetime]['Temperature'+str(dayDiff)] = row['temperature'] 
    dictData[datetime]['Weather'+str(dayDiff)] = row['wt']     
dictData
```
